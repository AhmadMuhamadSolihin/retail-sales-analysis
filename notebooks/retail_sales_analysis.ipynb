#Persiapan dan Cleaning Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Load Dataset

```python
df = pd.read_csv("supermarket_sales.csv")
df.head()
```

Melihat Struktur Dataset

```python
df.info()
```

```python
df.isna().sum()
```

Struktur dataset menunjukkan bahwa data memiliki 1000 baris dan 17 kolom (variabel). Dari keseluruhan variabel tersebut, tidak terdapat nilai kosong. Berdasarkan tipe datanya, dataset terdiri dari 7 variabel bertipe float64, 1 variabel bertipe int64, dan 9 variabel bertipe object. Variabel bertipe numerik seperti Unit price, Quantity, Tax 5%, Total, cogs, gross margin percentage, gross income, dan Rating merepresentasikan nilai kuantitatif yang berkaitan dengan transaksi penjualan dan keuntungan. Sementara itu, variabel bertipe object seperti Invoice ID, Branch, City, Customer type, Gender, Product line, Date, Time, dan Payment menggambarkan informasi kategorikal yang menjelaskan identitas transaksi, karakteristik pelanggan, jenis produk, serta metode pembayaran.

Mengubah Format Tanggal

```python
df['Date'] = pd.to_datetime(df['Date'])
```

Memisahkan Komponen Waktu

```python
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
```

```python
df['Hour'] = pd.to_datetime(df['Time']).dt.hour
```

```python
df['Sales_per_item'] = df['Unit price'] * df['Quantity']
```

Menyimpan data yang sudah dibersihkan

```python
df.to_csv("Data_Cleaned.csv", index=False)
```

#Explorasi Data Analys

##Produk Paling Laku

```python
df.groupby('Product line')['Quantity'].sum().sort_values(ascending=False)
```

Berdasarkan jumlah produk yang terjual pada setiap kategori, Electronic accessories merupakan produk dengan penjualan tertinggi dengan total 971 unit. Hal ini menunjukkan bahwa kategori tersebut merupakan produk paling diminati pelanggan dibandingkan kategori lainnya. Selanjutnya, kategori Food and beverages berada pada posisi kedua dengan 952 unit, diikuti oleh Sports and travel dengan 920 unit.
Secara keseluruhan, hasil ini menunjukkan bahwa produk yang berkaitan dengan **Electronic accessories** dan **kebutuhan konsumsi** memiliki tingkat permintaan yang lebih tinggi dibandingkan kategori produk lainnya.

##Produk dengan Keuntungan Tertinggi

```python
df.groupby('Product line')['gross income'].sum().sort_values(ascending=False)
```

Berdasarkan total gross income pada setiap kategori produk, Food and beverages merupakan kategori yang memberikan keuntungan terbesar dengan nilai sekitar 2673.56. Hal ini menunjukkan bahwa produk pada kategori tersebut memberikan kontribusi keuntungan paling tinggi bagi perusahaan dibandingkan kategori lainnya. Selanjutnya, kategori Sports and travel menempati posisi kedua dengan keuntungan sekitar 2624.90, diikuti oleh Electronic accessories sebesar 2587.50.
Secara keseluruhan, hasil ini menunjukkan bahwa meskipun beberapa kategori memiliki jumlah penjualan tinggi, kategori **Food and beverages** memberikan kontribusi keuntungan yang paling besar terhadap pendapatan perusahaan.


##Performa penjualan berdasarkan lokasi toko

```python
df.groupby('Branch')['Total'].sum()
```

```python
df.groupby('City')['Total'].sum()
```

Berdasarkan total penjualan pada setiap cabang, Branch C memiliki performa penjualan terbaik dengan total penjualan sekitar 110.568,71. Naypyitaw menjadi kota dengan total penjualan tertinggi yaitu sekitar 110.568,71, diikuti oleh Yangon dengan 106.200,37, dan Mandalay dengan 106.197,67.

Secara keseluruhan, dapat disimpulkan bahwa cabang yang berada di **Naypyitaw** memiliki performa penjualan paling baik, sedangkan cabang di Yangon dan Mandalay memiliki tingkat penjualan yang relatif seimbang.

##Segmentasi Pelanggan

```python
df.groupby('Customer type')['Total'].mean()
```

```python
df['Payment'].value_counts()
```

Berdasarkan segmentasi pelanggan, rata-rata nilai transaksi pelanggan member sedikit lebih tinggi dibandingkan pelanggan normal. Pelanggan member memiliki rata-rata total transaksi sekitar 327.79, sedangkan pelanggan normal memiliki rata-rata transaksi sekitar 318.12. Hal ini menunjukkan bahwa pelanggan member cenderung melakukan pembelian dengan nilai transaksi yang lebih besar. Selain itu, jika dilihat dari metode pembayaran, pelanggan paling banyak menggunakan E-wallet dengan 345 transaksi, diikuti oleh cash sebanyak 344 transaksi, dan credit card sebanyak 311 transaksi. Hal ini menunjukkan bahwa metode pembayaran digital dan tunai merupakan pilihan utama pelanggan dalam melakukan transaksi.

Secara keseluruhan, pelanggan member memberikan kontribusi transaksi yang sedikit lebih tinggi, sementara metode pembayaran E-wallet menjadi metode yang paling sering digunakan oleh pelanggan.

##Pola Waktu Tranksaksi

```python
df.groupby('Hour')['Total'].sum()
```

Berdasarkan pola waktu transaksi, penjualan terjadi pada rentang waktu pukul 10.00 hingga 20.00. Total transaksi tertinggi terjadi pada pukul 19.00 dengan nilai sekitar 39,699.51, yang menunjukkan bahwa aktivitas pembelian pelanggan paling ramai terjadi pada malam hari.

Selain itu, transaksi juga cukup tinggi pada pukul 13.00 dengan total sekitar 34,723.23, serta pada pukul 10.00 dan 15.00 yang masing-masing mencapai lebih dari 31,000. Sementara itu, aktivitas transaksi relatif lebih rendah pada pukul 20.00 dengan total sekitar 22,969.53.

Secara keseluruhan, pola ini menunjukkan bahwa waktu sore hingga malam hari merupakan periode dengan aktivitas transaksi paling tinggi, sehingga waktu tersebut menjadi momen penting bagi toko untuk mengoptimalkan pelayanan dan strategi penjualan.


##Tingkat Kepuasan Pelanggan

```python
df.groupby('Branch')['Rating'].mean()
```

Berdasarkan rata-rata rating pelanggan pada setiap cabang, secara umum tingkat kepuasan pelanggan berada pada nilai sekitar **7 dari 10**, yang menunjukkan bahwa pengalaman berbelanja pelanggan tergolong cukup baik.

Di antara ketiga cabang, **Branch B** memiliki rata-rata rating paling rendah yaitu sekitar **6.82**.

Hasil ini menunjukkan bahwa **Branch B masih memiliki peluang untuk meningkatkan kualitas layanan agar kepuasan pelanggan dapat lebih optimal**.


#Strategi Bisnis

Dari berbagai insight yang diperoleh, dapat dirumuskan beberapa strategi yang dapat diterapkan untuk mendorong pertumbuhan serta mengoptimalkan kinerja bisnis:


1.  **Optimalisasi Portofolio Produk**: Manfaatkan popularitas "Electronic accessories" sebagai penarik pelanggan, dan secara strategis dorong penjualan "Food and beverages" yang memiliki margin keuntungan lebih tinggi melalui bundling atau penempatan produk yang cerdas.

2.  **Replikasi Keberhasilan Cabang**: Pelajari secara mendalam faktor-faktor kunci keberhasilan "Cabang C di Naypyitaw". Identifikasi praktik terbaik dalam operasional, pemasaran, dan layanan pelanggan yang dapat direplikasi ke "Cabang A di Yangon" dan "Cabang B di Mandalay" untuk meningkatkan performa penjualan mereka.

3.  **Penguatan Loyalitas Pelanggan**: Tingkatkan investasi dalam program keanggotaan untuk "pelanggan Member" guna mendorong nilai transaksi yang lebih tinggi.

4.  **Manajemen Operasional Berbasis Waktu**: Optimalkan penjadwalan staf, alokasi sumber daya, dan aktivitas promosi sesuai dengan "pola waktu transaksi puncak", terutama pada pukul 19.00 dan 13.00, untuk memaksimalkan pengalaman pelanggan dan potensi penjualan. Pertimbangkan penawaran khusus pada jam-jam tersebut.

5.  **Peningkatan Kualitas Layanan Berbasis Data**: Jadikan "Cabang B" sebagai fokus utama untuk program peningkatan kualitas layanan. Lakukan audit layanan, pelatihan staf, atau "mystery shopping" untuk mengidentifikasi dan mengatasi akar penyebab "rating kepuasan yang lebih rendah".
